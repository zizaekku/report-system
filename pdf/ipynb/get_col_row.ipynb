```python
import olefile
import pandas as pd
```

```python
f = olefile.OleFileIO('../file/report_template.hwp')
#PrvText 스트림 내의 내용을 읽기
encoded_text = f.openstream('PrvText').read() 
#인코딩된 텍스트를 UTF-16으로 디코딩
decoded_text = encoded_text.decode('UTF-16') 
```

```python
decoded_text
```

```python
items = decoded_text.replace('\r', '')
```

```python
items
```

```python
list_items = items.split('[ ')
```

```python
df_items = pd.DataFrame(list_items)
```

```python
df_items
```

```python
col_row = df_items[0].str.extract(r'([A-Z]{1,2}[0-9]{1,4})')
```

```python
col_row
```

```python
drop_na_col_row = col_row.dropna(axis=0)
```

```python
drop_na_col_row
```

```python
drop_na_col_row['col'] = drop_na_col_row[0].str.extract(r'([A-Z]{1,2})')
```

```python
drop_na_col_row
```

```python
drop_na_col_row['row'] = drop_na_col_row[0].str.extract(r'([0-9]{1,2})')
```

```python
drop_na_col_row = drop_na_col_row.drop_duplicates([0])
```

```python
drop_na_col_row
```
